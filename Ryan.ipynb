```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#Importing Packages
```

```python
deprivation = pd.read_csv("DeprivationIndex.csv")
#Assigning our csv as a dataframe

deprivation.head
```

```python
lone_parent_prop = deprivation["Lone Parent Ratio 2022 %"]
primary_only_prop = deprivation["Prop. Primary Education Only 2022 %"]
third_level_prop = deprivation["Proportion at Third Level Education 2016 %"]

#Selecting the Variables of Interest for Question 2. Now Combining them into one dataframe. Dropping NAs

depr2 = pd.DataFrame({"Lone Parent Proportion (%)": lone_parent_prop, 
                      "Primary Schooling Only (%)": primary_only_prop})

depr2 = depr2.dropna()

depr3 = pd.DataFrame({"Lone Parent Proportion (%)": lone_parent_prop, 
                      "Third Level Proportion (%)": third_level_prop})

depr3 = depr3.dropna()
```

```python
correlation = depr2["Lone Parent Proportion (%)"].corr(
    depr2["Primary Schooling Only (%)"])

print("Correlation Coefficient:", correlation)

correlation1 = depr3["Lone Parent Proportion (%)"].corr(
    depr3["Third Level Proportion (%)"])

print("Correlation Coefficient:", correlation1)

#Calculating Correlation Coefficient
```

```python
plt.figure(1) 

plt.scatter(depr2["Primary Schooling Only (%)"],
            depr2["Lone Parent Proportion (%)"],
            s = 2)

plt.title("Correlation Between Primary Only Proportion and Lone Parent Proportion")

plt.figure(2)

plt.scatter(depr3["Third Level Proportion (%)"],
            depr3["Lone Parent Proportion (%)"],
            s = 2)

plt.title("Correlation Between Third Level Education Proportion and Lone Parent Proportion")

plt.show()

#Making a scatter plot with small dots.
```
